```python
#실습 - 2행 2열의 그래프를 그려보자.
#1행 1열: sin() 그래프 
#1행 2열: cos() 그래프 (표시 형식:빨간색 사각형)
#2행 1열: tan() 그래프 (표시 형식:청록색 점)
#2행 2열: arctan() 그래프 (표시 형식:노란색 원)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)
arctan_y=np.arctan(x)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].plot(x, sin_y, 'b--')
axs[0, 0].set_title('sin_graph')
axs[0, 0].set_xlabel('x-label')
axs[0, 0].set_ylabel('y-label')

axs[0, 1].plot(x, cos_y, 'rs')
axs[0, 1].set_title('cos_graph')
axs[0, 1].set_xlabel('x-label')
axs[0, 1].set_ylabel('y-label')

axs[1, 0].plot(x, tan_y, 'co')
axs[1, 0].set_title('tan_graph')
axs[1, 0].set_xlabel('x-label')
axs[1, 0].set_ylabel('y-label')

axs[1, 1].plot(x, arctan_y, 'yo')
axs[1, 1].set_title('arctan_graph')
axs[1, 1].set_xlabel('x-label')
axs[1, 1].set_ylabel('y-label')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)
arctan_y=np.arctan(x)

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(x, sin_y, 'b--')
plt.title('sin_graph')

plt.subplot(2, 2, 2)
plt.plot(x, cos_y, 'r--')
plt.title('cos_graph')

plt.subplot(2, 2, 3)
plt.plot(x, tan_y, 'c--')
plt.title('tan_graph')

plt.subplot(2, 2, 4)
plt.plot(x, arctan_y, 'y--')
plt.title('arctan_graph')

plt.show()
```

```python
x = range(0,15)
y = [v*v for v in x]

fig, axs = plt.subplots(2,2, figsize=(10,10))
axs[0,0].plot(x, y, 'o')
axs[0,0].set_title('o_plot_graph')

axs[0,1].plot(x, y, '-')
axs[0,1].set_title('-_plot_graph')

axs[1,0].bar(x, y)
axs[1,0].set_title('bar_graph')

axs[1,1].hist(x, y)
axs[1,1].set_title('hist_graph')

plt.show()
```
